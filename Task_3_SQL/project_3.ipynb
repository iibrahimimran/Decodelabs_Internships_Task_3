# Import Libraries

```python
import sqlite3
import pandas as pd
```

# Load the CSV

```python
df = pd.read_csv("D:/Computing Stastic 2/python/Decodelab/cleaned_dataset_final.csv")

df.head()
```

# Create a SQLite Database

```python
conn = sqlite3.connect("sales.db")

df.to_sql(
    "sales",
    conn,
    if_exists="replace",
    index=False
)

print("Database Created Successfully!")
```

# See All Columns

```python
pd.read_sql("""
SELECT *
FROM sales
LIMIT 10;
""", conn)
```

# Display all records

```python
pd.read_sql("""
SELECT *
FROM sales;
""", conn)
```

# Display only Product and TotalPrice

```python
pd.read_sql("""
SELECT Product, TotalPrice
FROM sales;
""", conn)
```

# Products with price greater than 500

```python
pd.read_sql("""
SELECT *
FROM sales
WHERE TotalPrice > 500;
""", conn)
```

# Orders paid using Credit Card

```python
pd.read_sql("""
SELECT *
FROM sales
WHERE PaymentMethod='Credit Card';
""", conn)
```

# Sort by Highest Price

```python
pd.read_sql("""
SELECT *
FROM sales
ORDER BY TotalPrice DESC;
""", conn)
```

# Sort by Lowest Price

```python
pd.read_sql("""
SELECT *
FROM sales
ORDER BY TotalPrice ASC;
""", conn)
```

# Count Total Orders

```python
pd.read_sql("""
SELECT COUNT(*) AS TotalOrders
FROM sales;
""", conn)
```

# Average Price

```python
pd.read_sql("""
SELECT AVG(TotalPrice) AS AveragePrice
FROM sales;
""", conn)
```

# Maximum Price

```python
pd.read_sql("""
SELECT MAX(TotalPrice) AS MaximumPrice
FROM sales;
""", conn)
```

# Minimum Price

```python
pd.read_sql("""
SELECT MIN(TotalPrice) AS MinimumPrice
FROM sales;
""", conn)
```

# Total Revenue

```python
pd.read_sql("""
SELECT SUM(TotalPrice) AS TotalRevenue
FROM sales;
""", conn)
```

# Total Sales by Product

```python
pd.read_sql("""
SELECT Product,
       SUM(TotalPrice) AS TotalSales
FROM sales
GROUP BY Product
ORDER BY TotalSales DESC;
""", conn)
```

# Average Quantity by Product

```python
pd.read_sql("""
SELECT Product,
       AVG(Quantity) AS AverageQuantity
FROM sales
GROUP BY Product;
""", conn)
```

# Number of Orders by Payment Method

```python
pd.read_sql("""
SELECT PaymentMethod,
       COUNT(*) AS Orders
FROM sales
GROUP BY PaymentMethod;
""", conn)
```

# Number of Orders by Status

```python
pd.read_sql("""
SELECT OrderStatus,
       COUNT(*) AS TotalOrders
FROM sales
GROUP BY OrderStatus;
""", conn)
```

# Top 5 Best Selling Products

```python
pd.read_sql("""
SELECT Product,
       SUM(TotalPrice) AS Sales
FROM sales
GROUP BY Product
ORDER BY Sales DESC
LIMIT 5;
""", conn)
```

# Orders with Coupon

```python
pd.read_sql("""
SELECT *
FROM sales
WHERE CouponCode IS NOT NULL;
""", conn)
```

# Referral Source Statistics

```python
pd.read_sql("""
SELECT ReferralSource,
       COUNT(*) AS TotalOrders
FROM sales
GROUP BY ReferralSource;
""", conn)
```

# Customers Who Purchased More Than One Item

```python
pd.read_sql("""
SELECT CustomerID,
       SUM(Quantity) AS TotalItems
FROM sales
GROUP BY CustomerID
HAVING TotalItems > 1;
""", conn)
```

# Close Database

```python
conn.close()
```
